## Thresholding

Using personal and instructed algorithms for thresholding:
- color_thres() - Personal
- anotehr_thres() - Instructed

```python
# Defining Personal Threshold Function
def color_thres(image,  rgb_thres=(150,150,150)):
    shape = ([image.shape[0], image.shape[1], 1])
    canvas = np.zeros(shape)
    thrs = np.asarray(rgb_thres).tolist()
    for y in range(image.shape[0]):
        for x in range(image.shape[1]):
            if image[y, x, :].tolist() < thrs:
                canvas[y,x] = 0
            else:
                canvas[y,x] = 1
    return canvas
```

```python
# Using Personal Threshold Function and Plotting Output
canvas = color_thres(img, rgb_thres= (160,160,160))
plt.imshow(canvas, cmap='Greys_r')
plt.show()
```

```python
# Defining Thrshold Function from Instructions
def another_thres(image, rgb_thres=(0,0,0)):
    shape = ([image.shape[0], image.shape[1], 1])
    canvas = np.zeros(shape)
    thres_check = (img[:,:,0] > rgb_thres[0]) & (img[:,:,1] > rgb_thres[1]) & (img[:,:,2] > rgb_thres[2])
    canvas[thres_check] = 1
    return canvas
```

```python
# Using Instruction Thrshold and Plotting Output
val = another_thres(img, (160,160,160))
plt.imshow(val, cmap='Greys_r')
plt.show()
```
